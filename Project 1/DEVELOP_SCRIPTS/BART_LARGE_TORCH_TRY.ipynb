```python
from transformers import BartTokenizer, BartForSequenceClassification, AutoConfig, BartModel
import torch

tokenizer = BartTokenizer.from_pretrained("facebook/bart-large")
model = BartModel.from_pretrained(
        "facebook/bart-large", config=AutoConfig.from_pretrained("facebook/bart-large", output_attentions=True, output_hidden_states=True))
```

```python
model
```

```python
inputs = tokenizer("Hello, my dog is cute", return_tensors="pt")
labels = torch.tensor([[1, 1, 1]], dtype=torch.float)  # need dtype=float for BCEWithLogitsLoss
outputs = model(**inputs)
```
